```python
# === BLOQUE 1: IMPORTACIONES Y CONFIGURACIÓN ===
import os
import json
import math
import warnings

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

warnings.filterwarnings('ignore')

# Crear carpeta de salida
os.makedirs("outputs", exist_ok=True)

def ensure_outputs():
    os.makedirs('outputs', exist_ok=True)

def log(msg: str):
    print(f'[INFO] {msg}')

def parse_date_series(s: pd.Series) -> pd.Series:
    # 1er intento: parse estándar (YYYY-MM-DD y similares)
    d1 = pd.to_datetime(s, errors='coerce')
    # 2do intento: forzar dayfirst en lo no parseado
    mask = d1.isna()
    if mask.any():
        d2 = pd.to_datetime(s[mask], errors='coerce', dayfirst=True)
        d1[mask] = d2
    return d1


def week_floor(dt_series: pd.Series) -> pd.Series:
    return dt_series.dt.to_period('W-MON').dt.start_time

# CARGA Y LIMPIEZA DE DATOS ===

def load_data():
    # Asegurarse que todo esté en minúsculas (según tu estructura)
    data_path   = 'data/data.csv'
    stores_path = 'data/stores.csv'
    oil_path    = 'data/oil.csv'

    log(f'Cargando archivos:\n - {data_path}\n - {stores_path}\n - {oil_path}')

    df_data   = pd.read_csv(data_path)
    df_stores = pd.read_csv(stores_path)

    tmp = pd.read_csv(oil_path, header=None, dtype=str, engine='python')
    header_row = next(
        (i for i, r in tmp.iterrows()
         if ('date' in r.astype(str).str.lower().tolist())
         or ('fecha' in r.astype(str).str.lower().tolist())),
        None
    )
    df_oil = pd.read_csv(oil_path, header=header_row, engine='python') if header_row is not None else pd.read_csv(oil_path)

    # Normalización
    df_data.columns   = df_data.columns.str.strip().str.lower()
    df_stores.columns = df_stores.columns.str.strip().str.lower()
    df_oil.columns    = df_oil.columns.str.strip().str.lower().str.replace('fecha', 'date', regex=False)

    # Fechas
    df_data['date'] = parse_date_series(df_data['date'])
    df_oil['date']  = parse_date_series(df_oil['date'])
    df_oil = df_oil.dropna(subset=['date']).sort_values('date').reset_index(drop=True)

    # Imputación precio petróleo
    if 'dcoilwtico' in df_oil.columns:
        df_oil['dcoilwtico'] = pd.to_numeric(df_oil['dcoilwtico'], errors='coerce')
        df_oil['oil_price_interp'] = df_oil['dcoilwtico'].interpolate(method='linear', limit_direction='both')
    elif 'oil_price' in df_oil.columns:
        df_oil['oil_price'] = pd.to_numeric(df_oil['oil_price'], errors='coerce')
        df_oil['oil_price_interp'] = df_oil['oil_price'].interpolate(method='linear', limit_direction='both')

    return df_data, df_stores, df_oil

# === BLOQUE 3: FUNCIONES DE EXPLORACIÓN ===

def resumen_simple(df, nombre):
    with open(f"outputs/{nombre}_resumen.txt", "w", encoding="utf-8") as f:
        f.write(f"=== {nombre} ===\n")
        df.info(buf=f)
        f.write("\nNulos por columna:\n")
        f.write(str(df.isnull().sum()))
        f.write("\n\nÚnicos por columna:\n")
        f.write(str(df.nunique()))

def ventas_diarias(df_data: pd.DataFrame) -> pd.Series:
    df = df_data[['date', 'sales']].copy()
    df['date']  = parse_date_series(df['date'])
    df['sales'] = pd.to_numeric(df['sales'], errors='coerce')
    s = df.groupby('date', as_index=True)['sales'].sum().sort_index()
    # Forzar frecuencia diaria para rolling estable.
    # Si hay días sin ventas, decide política: 0 (cierre/ausencia) vs interpolación.
    s = s.asfreq('D').fillna(0)
    return s

def plot_promedio_ventas_anual(df_data: pd.DataFrame, agg: str = 'sum'):
    df = df_data[['date','sales']].copy()
    df['date']  = parse_date_series(df['date'])
    df['sales'] = pd.to_numeric(df['sales'], errors='coerce')
    df['year']  = df['date'].dt.year

    if agg == 'sum':
        ventas_anuales = df.groupby('year', as_index=False)['sales'].sum()
        titulo = 'Ventas Totales por Año'
        ylabel = 'Ventas (suma)'
    else:
        ventas_anuales = df.groupby('year', as_index=False)['sales'].mean()
        titulo = 'Promedio de Ventas por Año'
        ylabel = 'Ventas (promedio)'

    plt.figure(figsize=(10, 5))
    plt.bar(ventas_anuales['year'].astype(str), ventas_anuales['sales'])
    plt.title(titulo); plt.xlabel('Año'); plt.ylabel(ylabel)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    ensure_outputs(); plt.tight_layout()
    plt.savefig('outputs/promedio_ventas_anual.png'); plt.show()

    print(ventas_anuales)

def plot_year_avg(df_data: pd.DataFrame, window_days: int = 365):
    # --- preparar serie diaria (suma por día) ---
    df = df_data[['date', 'sales']].copy()
    df['date']  = parse_date_series(df['date'])
    df['sales'] = pd.to_numeric(df['sales'], errors='coerce')

    # suma por día (si hay múltiples filas por fecha/tienda)
    s = df.groupby('date', as_index=True)['sales'].sum().sort_index()
    # frecuencia diaria explícita; día sin registro -> 0
    s = s.asfreq('D').fillna(0)

    # --- promedio anual móvil (365 días, centrado) ---
    mov_avg = s.rolling(window=window_days,
                        center=True,
                        min_periods=window_days//2).mean()

    # mostrar en miles
    s_k = s / 1000.0
    mov_k = mov_avg / 1000.0

    # --- gráfico ---
    plt.figure(figsize=(22, 9))
    plt.scatter(s_k.index, s_k.values, s=12, alpha=0.45, label='Ventas diarias (miles)')
    plt.plot(mov_k.index, mov_k.values, linewidth=3, color='red',
             label=f'Promedio anual móvil ({window_days} días)')
    plt.title('Ventas promedio por año', fontsize=14)
    plt.xlabel('date')
    plt.ylabel('Ventas (miles)')
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.legend()
    ensure_outputs()
    plt.savefig('outputs/ventas_promedio_por_anio.png', dpi=150, bbox_inches='tight')
    plt.show()

df_data, df_stores, df_oil = load_data()

resumen_simple(df_data, "DATA")
resumen_simple(df_stores, "STORES")
resumen_simple(df_oil, "OIL")

plot_promedio_ventas_anual(df_data)
plot_year_avg(df_data)      
```

```python
# === BLOQUE 4: PANEL SEMANAL SIMPLE POR TIENDA (SIN RENOMBRES) ===
def construir_panel_semanal(df_data: pd.DataFrame) -> pd.DataFrame:
    """
    Requiere EXACTAMENTE estas columnas en df_data:
      'date' (fecha), 'sales' (numérica), 'store' (id tienda), 'onpromotion' (numérica).
    Devuelve un panel semanal por tienda:
      week, store, sales (suma), promo_rate (promedio semanal).
    """
    df = df_data[["date", "sales", "store", "onpromotion"]].copy()

    # Tipos mínimos
    df["date"] = parse_date_series(df["date"])
    df["sales"] = pd.to_numeric(df["sales"], errors="coerce")
    df["onpromotion"] = pd.to_numeric(df["onpromotion"], errors="coerce")

    # Limpieza
    df = df.dropna(subset=["date", "sales"]).copy()
    df["week"] = week_floor(df["date"])

    # Agregado semanal por tienda
    wk = (
        df.groupby(["week", "store"], as_index=False)
          .agg(sales=("sales", "sum"),
               promo_rate=("onpromotion", "mean"))
    )

    ensure_outputs()
    wk.to_csv("outputs/weekly_panel_store.csv", index=False)
    return wk
```

```python
# === BLOQUE 5: SPLIT POR RANGOS DE FECHAS (TRAIN / VALID / TEST) ===
def split_por_fechas(wk: pd.DataFrame,
                     valid_inicio: str, valid_fin: str,
                     test_inicio: str,  test_fin: str):
    """
    RANGOS explícitos (YYYY-MM-DD) aplicados sobre 'week' (inicio de semana).
      TRAIN: week <  valid_inicio
      VALID: valid_inicio <= week <= valid_fin
      TEST : test_inicio  <= week <= test_fin
    """
    v_ini = pd.to_datetime(valid_inicio)
    v_fin = pd.to_datetime(valid_fin)
    t_ini = pd.to_datetime(test_inicio)
    t_fin = pd.to_datetime(test_fin)

    train = wk[wk["week"] < v_ini].copy()
    valid = wk[(wk["week"] >= v_ini) & (wk["week"] <= v_fin)].copy()
    test  = wk[(wk["week"] >= t_ini) & (wk["week"] <= t_fin)].copy()

    ensure_outputs()
    train.to_csv("outputs/train_panel_sem.csv", index=False)
    valid.to_csv("outputs/valid_panel_sem.csv", index=False)
    test.to_csv("outputs/test_panel_sem.csv", index=False)

    print("[SPLIT] TRAIN semanas:", train["week"].nunique(),
          "| VALID:", valid["week"].nunique(),
          "| TEST:",  test["week"].nunique())
    if len(valid):
        print("[VALID] ", str(valid["week"].min().date()), "→", str(valid["week"].max().date()))
    if len(test):
        print("[TEST ] ", str(test["week"].min().date()),  "→", str(test["week"].max().date()))
    return train, valid, test
```

```python
# === BLOQUE 6: USO MÍNIMO (AJUSTA LAS FECHAS) ===
wk_panel = construir_panel_semanal(df_data)

# EJEMPLO: cambia estos cortes por los tuyos (semanas comienzan en lunes)
train, valid, test = split_por_fechas(
    wk_panel,
    valid_inicio="2023-01-02", valid_fin="2023-07-03",
    test_inicio="2023-07-10",  test_fin="2023-12-25"
)
```
